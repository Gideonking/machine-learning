```python
import pandas as pd
import numpy as np
df = pd.read_csv('Human_Activity_Recognition_Using_Smartphones_Data.csv')
df['Activity'] = df['Activity'].map(
    {
        'STANDING':0,
        'SITTING':1,
        'LAYING':2,
        'WALKING':3,
        'WALKING_DOWNSTAIRS':4,
        'WALKING_UPSTAIRS':5,
    }
)
df.head()
```

```python
X = np.array(df.drop(['Activity'],1))
y = np.array(df['Activity'])
print(X.shape)
print(y.shape)
```

```python
from keras.layers import Dense, Activation, Flatten, Dropout
from keras.optimizers import Adam
from keras.utils import np_utils
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from keras import regularizers
# from keras.model_selection import train_test_split

labels = np_utils.to_categorical(y,6)
X = X.astype('float32')
y = y.astype('float32')



# X_train, x_test, Y_train, y_test = train_test_split(X,test_size = 0.3)
```

```python
model = Sequential([
    Dense(units=64, input_dim = 561),
    Activation('relu'),
    
    Dense(units = 32),
    Activation('relu'),
    
    Dense(units = 32),
    Activation('relu'),

    Dense(units = 6),
    Activation('softmax'),
])
```

model.summary()

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(X,labels,epochs=20,validation_split=0.2, batch_size = 25)
```
