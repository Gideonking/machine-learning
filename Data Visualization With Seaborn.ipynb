### Data Visualization With Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')

sns.lmplot(x = 'total_bill', y = 'tip',hue='sex', data = tips, palette='Set1')
plt.show()
```

### Making Sub-plots with col =.... (takes a column to divide as subplots)

```python
sns.lmplot(x = 'total_bill', y = 'tip',hue='sex',palette='Set1', data = tips, col='sex')
plt.show()
```

### Strip Plots

```python
sns.stripplot(y = 'total_bill', data=tips)
plt.ylabel('Tip ($)')
plt.show()
```

### Grouping with stripplots()

```python
sns.stripplot(x = 'day',y = 'tip', data = tips)
plt.ylabel('tip ($)')
plt.show()
```

### Grouping with stripplots (Spreading out strips with 'jitter' attribute)

```python
sns.stripplot(x = 'day',y = 'tip', data = tips, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

### BoxPlots and Violin Plots

```python
plt.subplot(1,2,1)
sns.boxplot(x = 'day', y = 'tip', data = tips)
plt.ylabel('tip ($)')


plt.subplot(1,2,2)
sns.violinplot(x = 'day', y = 'tip', data=tips)
plt.ylabel('tip ($)')

plt.tight_layout()
plt.show()
```
